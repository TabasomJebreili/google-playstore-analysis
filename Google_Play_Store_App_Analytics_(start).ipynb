## Introduction

This project analyzes the Android app ecosystem using publicly scraped data from the Google Play Store.  
We explore trends across categories, ratings, installs, pricing, and user reviews to identify patterns and insights in app performance.
The dataset includes metadata and user reviews for thousands of Android apps, scraped in 2018 by Lavanya Gupta.  
Source: [Google Play Store Apps on Kaggle](https://www.kaggle.com/lava18/google-play-store-apps)

# Numeric Type Conversion: Examine the Number of Installs


```python
df_apps_clean.head()
```

```python
type(df_apps_clean["Installs"].iloc[0])
```

```python
df_apps_clean.Installs = df_apps_clean.Installs.astype(str).str.replace(",", "")
```

## Top Paid Apps and Revenue Estimates

Estimate app revenue by multiplying price with install count, and identify the top 10 highest-grossing paid apps.  
Analyze how many of these are games versus other categories.


```python
type(df_apps_clean.Price.iloc[0])
```

```python
df_apps_clean.Price = df_apps_clean.Price.astype(str).str.replace('$', "")
```

```python
df_apps_clean.Price = pd.to_numeric(df_apps_clean.Price)
```

```python
df_apps_clean.sort_values('Price', ascending=False).head(20)
```

```python
df_apps_clean.drop([3946, 2394, 2461, 1331, 3145, 2775, 5765, 3221, 3554, 3114, 4606, 1946, 3897, 2193, 3856 ], inplace=True)
```

```python
revenue_estimate = df_apps_clean.Price * df_apps_clean.Installs
```

```python
df_apps_clean.insert(10, "Revenue", revenue_estimate)
```

# Plotly Bar Charts & Scatter Plots: Analysing App Categories

```python
app_category = df_apps_clean.groupby("Category").agg({
    "App": "count"
})
```

```python
top10_category = df_apps_clean.Category.value_counts()[:10]
top10_category
```

```python
bar = px.bar(x=top10_category.index, y=top10_category.values)
bar.show()
```

```python
category_installs = df_apps_clean.groupby('Category').agg({'Installs':'sum'})
```

```python
category_installs.head()
```

```python
category_installs.sort_values("Installs", ascending=True, inplace=True)
```

### Vertical Bar Chart - Highest Competition (Number of Apps)

```python
h_bar = px.bar(x=category_installs.Installs, y=category_installs.index, orientation='h', title="Category Popularity")
```

### Horizontal Bar Chart - Most Popular Categories (Highest Downloads)

```python
h_bar.update_layout(xaxis_title="Number of Downloads", yaxis_title="Category")
```

## Category Saturation: Installs vs. Competition

Visualize the relationship between category size (number of apps) and total installs using a Plotly scatter plot.  
The chart helps reveal which categories are overcrowded and which attract the most users.


```python
category_installs.head()
```

```python
concentration_df = pd.merge(app_category, category_installs, on="Category", how="inner")
```

```python
concentration_df.sort_values("Installs", ascending=False)
```

```python
scatter_fig = px.scatter(concentration_df, x="App", y="Installs", hover_name=concentration_df.index, size="App", color="Installs")
scatter_fig.show()
```

```python
scatter_fig.update_layout(xaxis_title="Number of Apps (Lower=More Concentrated)",
    yaxis_title="Installs",yaxis=dict(type='log'))
```

# Extracting Nested Data from a Column



```python
df_apps_clean.sample(5)
```

```python
df_apps_clean["Genres"].nunique()
```

```python
stack = df_apps_clean.Genres.astype(str).str.split(";", expand=True).stack()
```

## Visualizing Genre Popularity


```python
genre_bar = px.bar(x=number_of_genres.index[:15], y=number_of_genres.values[:15], title="Top Genres", color=number_of_genres.values[:15], color_continuous_scale='Agsunset')
```

```python
genre_bar.update_layout(xaxis_title="Genre", yaxis_title="Number of Apps", coloraxis_showscale=False)
```

# Plotly Box Plots: Revenue by App Category
